```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import nxviz as nv
from nxviz import MatrixPlot
from nxviz import CircosPlot
```

```python
g = nx.Graph()
```

```python
g.add_nodes_from([1,2,3])
```

```python
g.nodes()
```

```python
g.add_edge(1,2)
```

```python
g.edges()
```

```python
# can add metadata as well
g.node[1]['label'] = 'blue'
```

```python
g.nodes(data = True)
```

```python
nx.draw(g)
```

```python
G1 = nx.Graph()
G1.add_nodes_from([1,2,3,4,5])
G1.add_edges_from([(1,2), (1,3), (2,4), (3,5), (1,5)])
```

```python
ap = nv.ArcPlot(G1)
ap.draw()
plt.show()
```

```python
mp = MatrixPlot(G1)
mp.draw()
plt.show()
```

```python
cp = CircosPlot(G1)
cp.draw()
plt.show()
```

```python
# Find neighbors
list(G1.neighbors(2))
```

```python
# Find degree centrality
nx.degree_centrality(G1)
```

```python
nx.betweenness_centrality(G1)
```

```python
G = nx.barbell_graph(m1 = 5, m2 = 1)
nx.draw(G, with_labels = True)
```

```python
nx.find_cliques(G)
list(nx.find_cliques(G))
```
